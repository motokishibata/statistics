```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 平均
loc = 100
# 標準偏差
scale = 3

p1 = norm.cdf(x=90, loc=loc, scale=scale)
print(p1)

# 左端 (平均-5*σ)
start = loc - scale * 5
# 右端 (平均+5*σ)
end = loc + scale * 5

# X軸
X = np.arange(start, end, 0.1)
# 正規分布pdf生成
Y = norm.pdf(X, loc=loc, scale=scale)

# プロット
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid(color='gray')
ax.plot(X, Y, color='blue')
plt.show()
```
