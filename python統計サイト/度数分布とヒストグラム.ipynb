```python
import pandas as pd
import numpy as np

df = pd.read_csv('iris.csv')
data = np.array(df['sepal.length'])
hist, bin_edges = np.histogram(data, bins=10)
```

```python
hist_df = pd.DataFrame(columns=['start', 'end', 'count'])
for idx, val in enumerate(hist):
    start = round(bin_edges[idx], 2)
    end = round(bin_edges[idx + 1], 2)
    hist_df.loc[idx] = [start, end, val]
hist_df
```

上記の結果のように、データを区間分けしてそれぞれの数を集計したものを度数分布と呼ぶ。

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(data, bins=10, histtype='barstacked', ec='black')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(data, bins=[4, 5, 6, 7, 8], histtype='barstacked', ec='black')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(data, bins=[4, 5, 6, 7, 8], density=True,histtype='barstacked', ec='black')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(data, bins='auto', density=True,histtype='barstacked', ec='black')
plt.show()
```
