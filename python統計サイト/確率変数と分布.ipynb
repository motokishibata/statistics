サイコロの目のように各変数に対して発生する**確率が与えられているもの**を確率変数と呼ぶ。

|確率変数|確率|
|---|---|
|1|1/6|
|2|1/6|
|3|1/6|
|4|1/6|
|5|1/6|
|6|1/6|

サイコロの目は飛び飛びの値(温度などは連続した値)となっており、このような確率変数を離散型確率変数と呼ぶ。

```python
# サイコロの確率質量関数を作成して棒グラフを作成する

# 確率質量関数
def pmf(val):
    if (1 <= val and val <= 6):
        return 1 / 6
    return 0

import numpy as np
import matplotlib.pyplot as plt

left = [0, 1, 2, 3, 4, 5, 6, 7]
height = []
for i, v in enumerate(left):
    height.append(pmf(v))

ax = plt.axes()
ax.set_axisbelow(True)
ax.grid()

plt.xlabel("dice roll")
plt.ylabel("probability")
plt.bar(left, height, width=0.1)
plt.ylim(0, 0.3)
plt.show()
```

角度のような連続的な値を取る確率変数を連続型確率変数と呼ぶ。

```python
import math
import matplotlib.pyplot as plt

# 確率密度関数
def probabilityDensityFunction(x):
    if ((-math.pi) <= x and x < math.pi):
        return 1 / (2 * math.pi)
    return 0

left = [-5, -3, -1, 1, 3, 5]
height = []
for x in left:
    p = probabilityDensityFunction(x)
    print(p)
    height.append(p)

#角度の確率変数 -π～π
tmp = height[0]
x1 = 0
toggle = True
for (i, v) in enumerate(height):
    # 値が切り替わる x まで横線を引き、切り替わった x では縦線を引く
    if tmp != v:
        if toggle:
            plt.hlines(tmp, left[x1], left[i], color="green")
            plt.vlines(left[i], tmp, v, color="green")
        else:
            plt.hlines(tmp, left[x1], left[i - 1], color="green")
            plt.vlines(left[i - 1], tmp, v, color="green")

        x1 = i
        tmp = v
        toggle = (not toggle)
    
    if i == len(height) - 1:
        plt.hlines(v, left[x1 - 1], left[i], color="green")

# 0～90度の間に命中する確率
x_end = math.pi / 2
y_max = max(height)
plt.fill([0, 0, x_end, x_end], [0, y_max, y_max, 0], color="blue")

plt.grid()
plt.show()
```
