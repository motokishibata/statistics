```python
import pandas as pd
df = pd.read_csv('iris.csv')
```

```python
import numpy as np
data = np.array(df['sepal.length'])
np.mean(data)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(data, bins=10, histtype='barstacked', ec='black')
plt.show()
```
