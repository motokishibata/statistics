```python
import matplotlib.pyplot as plt

radio1 = [40, 35, 30, 25, 30]
radio2 = [50, 30, 60, 25, 65]

plt.plot(radio1)
plt.plot(radio2)
```

```python
from statistics import mean

avg1 = mean(radio1)
avg2 = mean(radio2)

tmp1 = 0
for r1 in radio1:
    diff = r1 - avg1
    tmp1 += diff * diff

d1 = tmp1 / len(radio1)

tmp2 = 0
for r2 in radio2:
    diff = r2 - avg2
    tmp2 += diff * diff

d2 = tmp2 / len(radio2)

print("分散1：", d1, " 分散2：", d2)
```

```python
print("標準偏差1：", d1**0.5, " 標準偏差2：", d2**0.5)
```
